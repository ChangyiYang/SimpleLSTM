# Data Checking

```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("./Processed Data/Cs_testing.csv")

def plot_line_chart(df, index1, index2):

    data1 = df.iloc[index1]
    data2 = df.iloc[index2]

    print(f"THe first P and T is {data1[1]} and {data1[2]}")
    print(f"THe second P and T is {data2[1]} and {data2[2]}")

    plt.plot(data1[3:], label=index1, alpha = 0.5)
    plt.plot(data2[3:], label=index2, alpha = 0.5)

    plt.legend()


    plt.show()

plot_line_chart(df, 100, 200)
```
