```python
import pandas as pd
from collections import defaultdict


training_data = pd.read_csv("./Processed Data/Cs_training.csv")
testing_data = pd.read_csv("./Processed Data/Cs_testing.csv")


data_dict = defaultdict(pd.DataFrame)




for _, row in training_data.iterrows():
    key = (row[1], row[2])

    data_dict[key] = data_dict[key].append(row[3:], ignore_index = True)

for _, row in testing_data.iterrows():
    key = (row[1], row[2])

    data_dict[key] = data_dict[key].append(row[3:], ignore_index = True)


print(data_dict.keys())

for key in data_dict.keys():
    print(data_dict[key].shape)
```

```python
import matplotlib.pyplot as plt
for key in data_dict.keys():
    plt.figure()
    plt.title((f"P:{key[0]/10000000}", f"T:{key[1]/72}"))
    for _, row in data_dict[key].iterrows():
        plt.plot(row, c = 'k', alpha = 0.01)
```

```python
raise KeyError
import pandas as pd
from collections import defaultdict
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set()

training_data = pd.read_csv("./Processed Data/Cs_training.csv")
testing_data = pd.read_csv("./Processed Data/Cs_testing.csv")

data_dict = defaultdict(pd.DataFrame)

for _, row in training_data.iterrows():
    key = (row[1], row[2])
    data_dict[key] = data_dict[key].append(row[3:], ignore_index=True)

for _, row in testing_data.iterrows():
    key = (row[1], row[2])
    data_dict[key] = data_dict[key].append(row[3:], ignore_index=True)

# Set the number of subplots
n = len(data_dict.keys())
fig, axs = plt.subplots(n, figsize=(15, n*5))

for i, key in enumerate(data_dict.keys()):
    for _, row in data_dict[key].iterrows():
        axs[i].plot(row, alpha=0.5, label=f"Key: {key}")
        axs[i].set_title(f"Key: {key}")
        axs[i].set_xlabel('Index')
        axs[i].set_ylabel('Value')
        axs[i].legend()

# Adjust layout
plt.tight_layout()
plt.show()
```
