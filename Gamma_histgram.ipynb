```python
# some imports

import pandas as pd   
import numpy as np
import pickle
import matplotlib.pyplot as plt
import random
```

```python
# do some peeeking

data = pd.read_pickle("./Gamma Data/gammas_detected_0.pkl")

table = data.iloc[0] #.unstack().T
```

```python
n_bins = 10

processed_data = []           

for time in  range(data.shape[0]):
    table = data.iloc[time]
    table = table.unstack()
    one_timestamp = []

    for  channel_index, one_channel in table.iterrows():
    # IMPORTANT: PLEASE REVIEW BEFORE PROCEEDING
    # ******************************************
    # the parameters in the np.histogramed function will greatly influence the final data processed result
    # important parameters
    # bins 
    # range 
    # ******************************************
    
    # hist, _ = np.histogram(one_channel, bins = 500, range = (0, 7200000))
        one_channel = one_channel.dropna()
        hist, _ = np.histogram(one_channel, bins = n_bins)
        if random.random() < 0.00001:
            plt.hist(one_channel, bins = n_bins)

        one_timestamp.append(hist)

    one_timestamp = np.concatenate(one_timestamp)

    processed_data.append(one_timestamp)


print(len(processed_data))
print(processed_data[0].shape)



processed_data = np.array(processed_data)


print(processed_data.shape)



```

```python
np.savetxt("./Gamma Data/Processed_cycle_32_bins_10.csv", processed_data, delimiter = ",")
```
